## MATPLOTLIB
plotting library, similar to MATLAB. Very well works with pandas, numpy. 
Seaborn is built on MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline #help us to see the graph in the notebook. 
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
print(x,y,sep='\n\n')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y label')
plt.title('Title')
# plt.show() #printing out the plot like in print('string')
```

### Multi Plot in same Canvas

```python
plt.subplot(1,2,1) #row, column, index
plt.plot(x,y,'r')

plt.subplot(1,2,2) # row column index 
plt.plot(y,x,'b')
```

### Object oriented method. 

```python
fig = plt.figure() #imaginary blank canvas 

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left right width height passed thru list. 

axes.plot(x,y) # we have more control over figure. 
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #between zero and one
axes2 = fig.add_axes([0.25,0.4,0.3,0.3]) #20,50,40,30 % of the corners. 
axes3 = fig.add_axes([0.65,0.2,0.2,0.2])

axes1.set_title("TP OTC ML Eng")

axes3.set_xlabel('Good Morning')
axes3.set_ylabel('19 TM')
axes3.set_title('Test')

axes2.set_xlabel('Drink Water, 10 Hello')
axes2.set_ylabel('11 Eat')
axes2.set_title('Just Write')

axes2.plot(x,y)
axes3.plot(y,x)
axes1.plot(x,y)
```

##### Fundamentals of OOP 

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.6,0.6])
axes1.plot(x,y)
```

### SubPlot with OOP Method

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()  #prevent from overlapping/

# for ax in axes:
#     ax.plot(x,y)

#alternatively we can use axes array to specify each axes. 
axes[0].plot(y,x)
axes[0].set_title('First Plot')

axes[1].plot(x,y)
axes[1].set_title('Second Plot')
```

#### Figure Size and DPI

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.text(3, 5, r"$y=x^2$", fontsize=10, color="blue",)
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title('Mola Ver')

axes[1].plot(y,x)
axes[1].set_title('Baris, Firat -> Ece Aykut Izmir')
plt.tight_layout()
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=3, alpha=0.5, ls=':',marker='o',ms=20, mfc='yellow')
ax.plot(x,x*3,color='blue',lw=2, ls='--',marker='+',markersize=10)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Plot Range

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.plot(x,y, ls=':',lw=3,color='brown')

axes.set_xlim([0,1]) #lower bound and upper bound 
axes.set_ylim([0,1.1])
```
