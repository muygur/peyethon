```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
#today we will focus very famous dataset called iris dataset.
```

```python
sns.pairplot(iris) #automated joint plot
```

```python
# we want more control over our plots.
g = sns.PairGrid(iris) #empty pair grid
g.map(plt.scatter) #scatter plot
```

```python
g = sns.PairGrid(iris) #empty pair grid
g.map_diag(sns.distplot) #diagonal plots
g.map_upper(plt.scatter) #upper plots
g.map_lower(sns.kdeplot) #lower plots
```

```python
# TIPS DATASET
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill','tip')
```
